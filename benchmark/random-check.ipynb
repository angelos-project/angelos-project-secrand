# SecureRandom Statistical Distributions
The statistical distributions of `Randomness` are implemented in `SecureRandom` class. Let's check the distributions by plotting histograms of random samples.


```kotlin
%use kandy
%use dataframe

import org.angproj.sec.SecureRandom
import org.jetbrains.kotlinx.dataframe.api.dataFrameOf

val dataframe = dataFrameOf(
    "Uniform" to List(10000) { SecureRandom.readUniform(0.0, 1.0) }
)

dataframe.plot {
    histogram("Uniform")
}
```

```kotlin
val dataframe = dataFrameOf(
    "Discrete Uniform" to List(10000) { SecureRandom.readDiscreteUniform(0, 100) }
)

dataframe.plot {
    histogram("Discrete Uniform")
}
```

```kotlin
val dataframe = dataFrameOf(
    "Gaussian" to List(10000) { SecureRandom.readGaussian() }
)

dataframe.plot {
    histogram("Gaussian")
}
```
