# Exponential Distribution
[https://en.wikipedia.org/wiki/Exponential_distribution](https://en.wikipedia.org/wiki/Exponential_distribution)

```kotlin
%use kandy
%use dataframe

import org.angproj.sec.SecureRandom
import org.jetbrains.kotlinx.dataframe.api.dataFrameOf

dataFrameOf(
    "a" to List(10000) { SecureRandom.readExponential(0.5) },
    "b" to List(10000) { SecureRandom.readExponential(1.0) },
    "c" to List(10000) { SecureRandom.readExponential(1.5) },
).plot {
    densityPlot("a") { alpha = 0.5; fillColor = Color.RED }
    densityPlot("b") { alpha = 0.5; fillColor = Color.GREEN }
    densityPlot("c") { alpha = 0.5; fillColor = Color.BLUE }
}
```
