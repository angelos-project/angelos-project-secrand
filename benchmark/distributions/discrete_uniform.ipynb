# Discrete Uniform Distribution
[https://en.wikipedia.org/wiki/Discrete_uniform_distribution](https://en.wikipedia.org/wiki/Discrete_uniform_distribution)

```kotlin
%use kandy
%use dataframe

import org.angproj.sec.SecureRandom
import org.jetbrains.kotlinx.dataframe.api.dataFrameOf

dataFrameOf(
    "Discrete Uniform" to List(10000) { SecureRandom.readDiscreteUniform(25, 75) }
).plot {
    histogram("Discrete Uniform")
}
```
