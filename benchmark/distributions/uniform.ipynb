# Uniform (Continuous) Distribution
[https://en.wikipedia.org/wiki/Continuous_uniform_distribution](https://en.wikipedia.org/wiki/Continuous_uniform_distribution)

```kotlin
%use kandy
%use dataframe

import org.angproj.sec.SecureRandom
import org.jetbrains.kotlinx.dataframe.api.dataFrameOf

dataFrameOf(
    "Standard" to List(10000) { SecureRandom.readUniform(0.25, 0.50) }
).plot {
    histogram("Standard")
}
```
