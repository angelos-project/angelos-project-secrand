# Gaussian Distribution
[https://en.wikipedia.org/wiki/Normal_distribution](https://en.wikipedia.org/wiki/Normal_distribution)

```kotlin
%use kandy
%use dataframe

import org.angproj.sec.SecureRandom
import org.jetbrains.kotlinx.dataframe.api.dataFrameOf

dataFrameOf(
    "a" to List(1000) { SecureRandom.readGaussian(0.0, 0.2) },
    "b" to List(1000) { SecureRandom.readGaussian(0.0, 1.0) },
    "c" to List(1000) { SecureRandom.readGaussian(0.0, 5.0) },
    "d" to List(1000) { SecureRandom.readGaussian(-2.0, 0.5) }
).plot {
    densityPlot("a") { alpha = 0.5; fillColor = Color.RED }
    densityPlot("b") { alpha = 0.5; fillColor = Color.GREEN }
    densityPlot("c") { alpha = 0.5; fillColor = Color.BLUE }
    densityPlot("d") { alpha = 0.5; fillColor = Color.YELLOW }
}
```
